```python
import pandas as pd
data = pd.read_csv('dataset/rating_final.csv')
data.head()
```

```python

import psycopg2
import numpy as np

conn = psycopg2.connect(database='reco_db', user='reco_user',password='reco_pass')

cur = conn.cursor()

create_sql = """
CREATE TABLE interactions(
    placeID BIGINT,
    userID TEXT,
    rating DOUBLE PRECISION,
    food_rating DOUBLE PRECISION,
    service_rating DOUBLE PRECISION,
    trained BOOLEAN DEFAULT TRUE,
    PRIMARY KEY (placeID, userID)
)
"""

cur.execute(create_sql)

insert_sql = """
INSERT INTO interactions (
    placeID, userID, rating, food_rating, service_rating
)
VALUES (
    %s,%s,%s,%s,%s
)
ON CONFLICT (userID, placeID) DO NOTHING
"""

for _, row in data.iterrows():
    cur.execute(
        insert_sql,
        (   
            int(row.placeID),
            row.userID,
            float(row.rating),
            float(row.food_rating),
            float(row.service_rating)
     
        )
    )

conn.commit()
```
